HashMap Program (Two Sum)

```python
nums=[2,6,7,11,14]
target=9
hashmap={}
for i in range(len(nums)):
    complement=target-nums[i]
    if complement in hashmap:
        print("Indices:", hashmap[complement],i)
        break
    hashmap[nums[i]]=i
```

```python
table={}
keys=[25,35,45]
for key in keys:
    index=key%10
    if index not in table:
        table[index]=[]
    table[index].append(key)
print(table)
```

Majority Element

```python
nums = [2, 2, 1, 2, 3]

count = {}

for num in nums:
    count[num] = count.get(num, 0) + 1

for key in count:
    if count[key] > len(nums) // 2:
        print("Majority Element:", key)
```

HashMap (dict) for O(1) Lookups

```python
# HashMap Lookup Example
items = [10, 20, 30, 20, 40]
data = {}
for item in items:

    if item in data:
        print(item, "already exists")

    else:
        data[item] = "Stored"
        print(item, "stored")

print("\nHashMap:")
print(data)
```

Another Example (Stores Frequency)

```python
items = [10, 20, 10, 30, 20, 10]
data = {}
for item in items:

    if item in data:
        data[item] += 1

    else:
        data[item] = 1
print(data)
```
