```python
# Hash Function Example

key = "Anu"

hash_value = hash(key)

print("Key:", key)
print("Hash Value:", hash_value)
```

HashMap Program (Two Sum)

```python
nums=[2,6,7,11,14]
target=9
hashmap={}
for i in range(len(nums)):
    complement=target-nums[i]
    if complement in hashmap:
        print("Indices:", hashmap[complement],i)
        break
    hashmap[nums[i]]=i
```
