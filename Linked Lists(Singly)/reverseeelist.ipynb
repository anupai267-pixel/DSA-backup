```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse(head):
    prev = None
    current = head

    while current:
        next = current.next
        current.next = prev
        prev = current
        current = next

    return prev

def display(head):
    while head:
        print(head.data, end=" -> ")
        head = head.next
    print("None")

# Create Linked List
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

print("Original List:")
display(head)

head = reverse(head)

print("Reversed List:")
display(head)

```
