Insertion at Head

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
head = None
# Insert 10
new_node = Node(10)
new_node.next = head
head = new_node
# Insert 20
new_node = Node(20)
new_node.next = head
head = new_node
# Print Linked List
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("None")
```

Insertion at Tail

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
head = Node(10)
new_node = Node(20)
temp = head

while temp.next:
    temp = temp.next

temp.next = new_node

# Print Linked List
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("None")
```

Insertion at middle

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
# Create Linked List: 10 -> 20 -> 40
head = Node(10)
second = Node(20)
third = Node(40)

head.next = second
second.next = third

# Insert 30 at Position 3
position = 3
new_node = Node(30)
temp = head

for i in range(position - 2):
    temp = temp.next
new_node.next = temp.next
temp.next = new_node

# Display Linked List
temp = head

while temp:
    print(temp.data, end=" -> ")
    temp = temp.next
print("None")
```
