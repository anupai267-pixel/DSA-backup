```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def detect(head):
    s = set()

    while head:
        if head in s:
            return "Cycle Found"
        s.add(head)
        head = head.next

    return "No Cycle"

# Create nodes
a = Node(1)
b = Node(2)
c = Node(3)

# Link nodes
a.next = b
b.next = c
c.next = b      # Creates a cycle

print(detect(a))
```
