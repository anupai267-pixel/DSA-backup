Deletion at Head

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Create Linked List
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

# Delete Head
head = head.next

# Display
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next

print("None")
```

Deletion at Middle

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Create Linked List
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

position = 3

temp = head

for i in range(position - 2):
    temp = temp.next

temp.next = temp.next.next

# Display
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next

print("None")
```

Deletion at Tail

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Create Linked List
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

temp = head

while temp.next.next:
    temp = temp.next

temp.next = None

# Display
temp = head
while temp:
    print(temp.data, end=" -> ")
    temp = temp.next

print("None")
```
