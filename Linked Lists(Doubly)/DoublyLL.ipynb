1. Traversal

```python
class Node:
    def __init__(self,data):
        self.prev=None
        self.data=data
        self.next=None
a = Node(10)
b = Node(20)
c = Node(30)

a.next = b
b.prev = a
b.next = c
c.prev = b

temp = a

while temp:
    print(temp.data)
    temp = temp.next
```

2. Insert at Beginning

```python
class Node:
    def __init__(self, data):
        self.prev = None
        self.data = data
        self.next = None

a = Node(20)
b = Node(30)

a.next = b
b.prev = a

new = Node(10)

new.next = a
a.prev = new
a = new

temp = a

while temp:
    print(temp.data)
    temp = temp.next
```
