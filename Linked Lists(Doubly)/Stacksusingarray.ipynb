```python
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print("After Push:", stack)

stack.pop()

print("After Pop :", stack)

print("Top Element:", stack[-1])
```

2. Stack Using Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

top = None

# Push 10
new = Node(10)
new.next = top
top = new

# Push 20
new = Node(20)
new.next = top
top = new

# Push 30
new = Node(30)
new.next = top
top = new

print("Top Element:", top.data)
```
