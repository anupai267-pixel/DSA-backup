```python
def nextGreater(arr):

    stack = []
    result = [-1] * len(arr)

    for i in range(len(arr)):

        while stack and arr[i] > arr[stack[-1]]:
            index = stack.pop()
            result[index] = arr[i]

        stack.append(i)

    return result


arr = [4, 5, 2, 10]

ans = nextGreater(arr)

for i in range(len(arr)):
    print(arr[i], "-->", ans[i])
```

```python
def nextGreater(arr):

    stack = []
    result = [-1] * len(arr)

    for i in range(len(arr)-1, -1, -1):

        while stack and arr[stack[-1]] <= arr[i]:
            stack.pop()

        if stack:
            result[i] = arr[stack[-1]]

        stack.append(i)

    return result


arr = [4, 5, 2, 10]

ans = nextGreater(arr)

for i in range(len(arr)):
    print(arr[i], "-->", ans[i])
```
