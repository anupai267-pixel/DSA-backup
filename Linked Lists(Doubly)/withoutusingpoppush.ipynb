```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
# Existing Stack
head = Node(20)
head.next = Node(10)
# Stack before Push
print("Before Push:")
temp = head
while temp:
    print(temp.data)
    temp = temp.next
# Push 30
new = Node(30)
new.next = head
head = new
# Stack after Push
print("After Push:")
temp = head
while temp:
    print(temp.data)
    temp = temp.next
```
