Python Collections Counter

```python
from collections import Counter
a=[1,1,1,2,2,2,3,3,4]
cnt=Counter(a)
print(cnt)
```

Creating a Counter

```python
from collections import Counter
ctr1=Counter([1,2,3,3,2,3])
ctr2=Counter({1:2,2:3,3:1})
ctr3=Counter('hello')
print(ctr1)
print(ctr2)
print(ctr3)
```

Accessing Counter Elements

```python
from collections import Counter
ctr = Counter([1,1,2,2,2,3])
print(ctr[1])
print(ctr[2])
print(ctr[3])
print(ctr[4])
```

Counter Methods

1. update(): 

```python
from collections import Counter
ctr=Counter([1,1,2])
ctr.update([2,2,3,3])
print(ctr)
```

2. elements(): 

```python
from collections import Counter
ctr =Counter([1,1,3,2,2,2,3])
items=list(ctr.elements())
print(items)
```

3. most_common():

```python
from collections import Counter
ctr=Counter([1,2,2,2,3,3,3,3,4])
common=ctr.most_common(1)
print(common)
```

4. Increasing Count Manually:

```python
from collections import Counter
ctr=Counter([1,1,1,2,3,3])
ctr[2]+=2
ctr[4]+=1
print(ctr)
```

5. subtract():

```python
from collections import Counter
ctr=Counter([1,1,1,2,2,2,2,3,3])
ctr.subtract([2,3])
print(ctr)
```

Arithmetic Operations on Counters

```python
from collections import Counter
ctr1=Counter([1,2,2,3,3])
ctr2=Counter([2,3,3,4])
print(ctr1 + ctr2)
print(ctr1 - ctr2)
print(ctr1 & ctr2)
print(ctr1 | ctr2)
```
