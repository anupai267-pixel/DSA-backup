numpy Arrays

```python
import numpy as np
a = np.array([1, 2, 3, 4])
print(a * 2)  
res = np.array([[1, 2], [3, 4]])
print(res * 2)
```

```python
import array as arr
a=arr.array('i',[1,2,3])
print(a[0])
a.append(5)
print(a)
```

```python
import array as arr
a=arr.array('i',[1,2,3])
print(*a)
a.insert(1,4)
print(*a)
a.remove(1)
print(a)
a.pop(2)
print(a)
res=a[1:2]
print(res)
```

Searching Element

Updating Elements

```python
import array
a = array.array('i', [1, 2, 3, 1, 2, 5])
a[2] = 6
print(a)
a[4] = 8
print(a)
```

```python
import array
a = array.array('i', [1, 2, 3, 4, 2, 5, 2])
count = a.count(2)
print(count)
a.reverse()
print(*a)
a.extend([6,7,8,9,10])
print(a)
```
