Abstraction

```python
from abc import ABC, abstractmethod

class Animal(ABC):      # Abstract Class

    @abstractmethod
    def sound(self):    # Abstract Method
        pass

    def sleep(self):    # Concrete Method
        print("Animal is sleeping")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
d.sleep()
```

Abstract Property Example


```python
from abc import ABC, abstractmethod
class Person(ABC):            #Abstract class
    @property
    @abstractmethod           
    def name(self):           #Abstract method
        pass
    def show(self):           # Concrete Method
        print("This is a person")
class Student(Person):
    @property                 #Abstract property
    def name(self):
        return "Anuradha"
s=Student()                   #Abstract Instantiation
print(s.name)
s.show()
```

What if the child class does not define the property?

```python
from abc import ABC, abstractmethod
class Person(ABC):
    @property
    @abstractmethod
    def name(self):
        pass
class Student(Person):
    pass
s = Student()
```
