Encapsulation

```python
class Student:
    def __init__(self):
        self.name="Anu"
    def display(self):
        print("Name:",self.name)
s = Student()
s.display()
```

Public Variable

```python
class Student:
    def __init__(self):
        self.name = "Anu"   #Public variable
s = Student()
print(s.name)
```

Protected Variable

```python
class Student:
    def __init__(self):
        self._name="anuradha"   #Protected variable
s=Student()
print(s._name)
```

Private Variable  ------cannot be accessed directly Hence ERROR

```python
class Student:
    def __init__(self):
        self.__name="Anuradha"     #Private variable
s=Student()
print(s.__name)
```

Private variable

```python
class Student:
    def __init__(self):
        self.__name="Anuradha"      #Private variable
    def display(self):
        print(self.__name)
s=Student()
s.display()
```

Getter Method

```python
class Student:
    def __init__(self):
        self.__name="Anu"
    def get_name(self):
        return self.__name
s=Student()
print(s.get_name())
```

Setter Method

```python
class Student:
    def __init__(self):
        self.__name="anu"
    def set_name(self,name):
        self.__name=name
    def get_name(self):
        return self.__name
s=Student()
s.set_name("Riya")
print(s.get_name())
```

```python
class Student:
    def __init__(self):
        self.__name = "Anu"   # Private variable

    def display(self):
        print(self.__name)

s = Student()
s.display()
```
