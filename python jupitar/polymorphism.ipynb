Compile-Time Polymorphism (static/Overloading)

```python
class Demo:
    def add(self,a,b=0,*arg):
        return a+b
obj = Demo()
print(obj.add(5))
print(obj.add(5,20))
print(obj.add(5,10,15))
```

Runtime Polymorphism (dynamic/Overriding)

```python
class Animal:
    def sound(self):
        return("Animal makes sound")
class Dog(Animal):
    def sound(self):
        return("Bark")
d=Dog()
print(d.sound())
```
