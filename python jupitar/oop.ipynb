Python OOP Concepts

Creating Class and Object

```python
class Dog:
    species="Canine"
    def __init__(self,name,age):
        self.name=name
        self.age=age
dog1=Dog("Buddy",3)
print(dog1.name)
print(dog1.species)
```

```python
class Dog:
    sound="bark"
dog1=Dog()
print(dog1.sound)
```

Initiate Object with __init__()

```python
class Dog:
    species = "Canine"
    def __init__(self,name,age):
        self.name=name
        self.age=age
dog1=Dog("Buddy",3)
print(dog1.name)
print(dog1.species)
```

__str__() Method

```python
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name} is {self.age} years old."
dog1=Dog("Buddy",3)
dog2=Dog("Charlie",5)
print(dog1)
print(dog2)
```
