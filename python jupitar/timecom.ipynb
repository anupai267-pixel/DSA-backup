O(1) – Constant Time

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[2])
```

O(n) – Linear Time

```python
numbers = [10, 20, 30, 40, 50]

for num in numbers:
    print(num)
```

O(log n) – Logarithmic Time

```python
numbers = [10, 20, 30, 40, 50, 60, 70]
target = 50

low = 0
high = len(numbers) - 1

while low <= high:
    mid = (low + high) // 2

    if numbers[mid] == target:
        print("Found")
        break

    elif numbers[mid] < target:
        low = mid + 1

    else:
        high = mid - 1
```

O(n²) – Quadratic Time

```python
n = 4

for i in range(n):
    for j in range(n):
        print(i, j)
```
