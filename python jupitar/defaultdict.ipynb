Defaultdict

```python
from collections import defaultdict
d = defaultdict(list)

d['fruits'].append('apple')
d['vegetables'].append('carrot')
print(d)
print(d['juices'])
```

Use Cases of defaultdict

1. Using List as Default Factory

```python
from collections import defaultdict
d = defaultdict(list)
for i in range(5):
    d[i].append(i)
    
print("Dictionary with values as list:")
print(d)
```

2. Using int Default Factory

```python
from collections import defaultdict
d=defaultdict(int)
a=[1,2,3,5,3,3,3,3,4,4,4,5,5,5,6]
for i in a:
    d[i] += 1
print(d)
```

3. Using str Default Factory

```python
from collections import defaultdict
sd=defaultdict(str)
sd['greeting']='Hello'
print(sd)
```

4. Grouping Words by First Letter

```python
from collections import defaultdict
words=["apple","ant","banana","bat"]
grouped=defaultdict(list)
for word in words:
    grouped[word[0]].append(word)
print(grouped)
```

Python defaultdict Type for Handling Missing Keys

```python
from collections import defaultdict
d = defaultdict(lambda: "Not Present")
d["a"] = 1
d["b"] = 2
print(d.__missing__('x'))
print(d.__missing__('d'))
print(d['a'])
```
