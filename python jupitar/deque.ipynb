Deque

```python
from collections import deque
d=deque(['name','age','DOB'])
print(d)
```

Accessing Items in deque

```python
from collections import deque
dq=deque([1,2,3,4])
print(dq[0])
print(dq[-1])
```

Operations on deque

1. append():

```python
from collections import deque
dq=deque([1,2,3,4])
dq.append(7)
print(dq)
```

2. appendleft():

```python
from collections import deque
dq=deque([1,3,4,5])
dq.appendleft(6)
print(dq)
```

3.extend()

```python
from collections import deque
dq=([1,2,3,4,5])
dq.extend([7,8,9])
print(dq)
```

4.extendleft() 

```python
from collections import deque
dq=deque([1,2,3,4,5])
dq.extendleft([8,9])
print(dq)
```

5.remove()

```python
from collections import deque
dq=deque([1,2,3,4])
dq.remove(3)
print(dq)
```

6. pop()

```python
from collections import deque
dq=deque([1,2,3,4,5])
dq.pop()
print(dq)
```

7. popleft()

```python
from collections import deque
dq=deque([1,2,3,4])
dq.popleft()
print(dq)
```

8. clear()

```python
from collections import deque
dq=deque([1,2,3,4,5])
dq.clear()
print(dq)
```

9.len()

```python
from collections import deque
dq=deque([1,2,3,4])
print(len(dq))
```

10. count():

```python
from collections import deque
dq=deque([1,2,3,4,4,4])
print(dq.count(4))
```

11. rotate(): 

```python
from collections import deque
dq = deque([10, 20, 30, 40])
dq.rotate(1)
print(dq)
```

12. reverse():

```python
from collections import deque
dq = deque([10, 20, 30, 40])
dq.reverse()
print(dq)
```
