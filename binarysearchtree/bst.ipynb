```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
    def insert(self, data):
        if data < self.data:
            if self.left is None:
                self.left = Node(data)
            else:
                self.left.insert(data)
        else:
            if self.right is None:
                self.right = Node(data)
            else:
                self.right.insert(data)
# List of values
values = [50, 30, 70, 20, 40, 60, 80]
# Create BST
root = Node(values[0])
# Insert remaining values
for value in values[1:]:
    root.insert(value)
# Print the tree nodes
print("Root =", root.data)
print("Left Child =", root.left.data)
print("Right Child =", root.right.data)
print("Left of 30 =", root.left.left.data)
print("Right of 30 =", root.left.right.data)
print("Left of 70 =", root.right.left.data)
print("Right of 70 =", root.right.right.data)
```

BST using insert search delete

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def insert(root, data):
    if root is None:
        return Node(data)

    if data < root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)

    return root


def search(root, key):
    if root is None:
        return False

    if root.data == key:
        return True

    if key < root.data:
        return search(root.left, key)

    return search(root.right, key)   # ✅ Fixed


def delete(root, key):
    if root is None:
        return None

    if key < root.data:
        root.left = delete(root.left, key)

    elif key > root.data:
        root.right = delete(root.right, key)

    else:
        # Delete node with 0 or 1 child
        if root.left is None:
            return root.right

        if root.right is None:
            return root.left

    return root


def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)


# Main
root = None

values = [50, 30, 70, 20, 40, 60]

for value in values:
    root = insert(root, value)

print("BST (Inorder):")
inorder(root)

print("\nSearch 40:", search(root, 40))

root = delete(root, 20)

print("After deleting 20:")
inorder(root)
```
