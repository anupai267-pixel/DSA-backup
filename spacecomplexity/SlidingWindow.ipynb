```python
arr=[2,1,5,1,3,2]
k=3
window_sum=sum(arr[:k])
max_sum=window_sum 
for i in range(k,len(arr)):
    window_sum=window_sum-arr[i-k]+arr[i]
    max_sum=max(max_sum,window_sum)
print("Maximum Sum=",max_sum)
```

Average of Every Subarray of Size K

```python
arr = [1, 3, 2, 6, -1, 4, 1, 8, 2]
k = 3

window_sum = sum(arr[:k])
print(window_sum / k)

for i in range(k, len(arr)):
    window_sum = window_sum - arr[i-k] + arr[i]
    print(window_sum / k)
```
