```python
def find_max(arr):
    max_num = arr[0]

    for i in arr:
        if i > max_num:
            max_num = i

    return max_num

arr = [5, 2, 9, 1]
print(find_max(arr))
```

Sum of Elements

```python
def find_sum(arr):
    total = 0
    for i in arr:
        total += i
    return total
print(find_sum([1,2,3,4]))
```

linear search

```python
def linear_search(arr,key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

print(linear_search([10, 20, 30, 40], 30))
```
