Copy Array

```python
def copy_array(arr):
    new_arr=[]
    for i in arr:
        new_arr.append(i)
    return new_arr
print(copy_array([1,2,3,4]))
```

Store Squares

```python
def squares(arr):
    result = []
    for i in arr:
        result.append(i)
    return result 
print(squares([1,2,3,4]))
```

Reverse Array

```python
def reverse_array(arr):
    rev = []
    for i in range(len(arr)-1, -1, -1):
        rev.append(arr[i])
    return rev
print(reverse_array([1,2,3,4]))
```
