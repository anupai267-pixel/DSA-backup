Height of a Binary Tree-DFS (Recursion) 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def height(root):
    if root is None:
        return -1      # Height in terms of edges

    return 1 + max(height(root.left), height(root.right))

# Tree
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.left = Node("D")
root.left.right = Node("E")

print("Height:", height(root))
```

height of tree - BFS (Level Order Traversal)

```python
from collections import deque

def height(root):

    if root is None:
        return 0

    queue = deque([root])

    height = 0

    while queue:

        size = len(queue)

        for i in range(size):

            node = queue.popleft()

            if node.left:
                queue.append(node.left)

            if node.right:
                queue.append(node.right)

        height += 1

    return height


print(height(root))
```
