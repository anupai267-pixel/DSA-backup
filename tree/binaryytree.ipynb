```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# Creating nodes
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.left = Node("D")
root.left.right = Node("E")

# Printing
print("Root:", root.data)
print("Left Child:", root.left.data)
print("Right Child:", root.right.data)
print("Leaf Nodes:", root.left.left.data, root.left.right.data, root.right.data)
```

Height of a Binary Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def height(root):
    if root is None:
        return -1      # Height in terms of edges

    return 1 + max(height(root.left), height(root.right))

# Tree
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.left = Node("D")
root.left.right = Node("E")

print("Height:", height(root))
```
