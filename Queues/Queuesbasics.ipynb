Simple Queue Program (Using List)

```python
queue = []

# Enqueue
queue.append(10)
queue.append(20)
queue.append(30)

print("Queue after Enqueue:", queue)

# Dequeue
if queue:
    removed = queue.pop(0)
    print("Removed element:", removed)

print("Queue after Dequeue:", queue)
```

```python
from collections import deque

dq = deque()

# Enqueue
dq.append(10)
dq.append(20)
dq.append(30)

print("Queue after Enqueue:", dq)

# Dequeue
removed = dq.popleft()

print("Removed element:", removed)
print("Queue after Dequeue:", dq)
```

```python
from collections import deque

dq = deque()

dq.append(10)      # Enqueue
dq.append(20)
print(dq)

dq.popleft()       # Dequeue
print(dq)
```

```python
from collections import deque

dq = deque()

# 1. Add at Rear (Enqueue)
dq.append(10)
dq.append(20)
print(dq)
# Output: deque([10, 20])

# 2. Add at Front
dq.appendleft(5)
print(dq)
# Output: deque([5, 10, 20])

# 3. Remove from Front (Dequeue)
x = dq.popleft()
print("Removed:", x)
print(dq)
# Output:
# Removed: 5
# deque([10, 20])

# 4. Remove from Rear
x = dq.pop()
print("Removed:", x)
print(dq)
# Output:
# Removed: 20
# deque([10])

# 5. Front Element
print("Front:", dq[0])
# Output: Front: 10

# 6. Rear Element
print("Rear:", dq[-1])
# Output: Rear: 10

# 7. Check if Empty
if not dq:
    print("Queue is Empty")
else:
    print("Queue is Not Empty")

# 8. Size of Queue
print("Size:", len(dq))
# Output: Size: 1

# 9. Clear Queue
dq.clear()
print(dq)
# Output: deque([])
```
