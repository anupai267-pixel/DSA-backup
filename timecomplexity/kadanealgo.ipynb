```python
def kadane(arr):
    current_sum = arr[0]
    max_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)

    return max_sum

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

result = kadane(arr)

print("Maximum Sum =", result)
```

```python
def kadane(arr):
    current_sum = arr[0]
    max_sum = arr[0]

    for num in arr[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

arr = [-2, -3, 4, -1, -2, 1, 5, -3]
arr1=[2,4,5,6,7,8]

print("Maximum Sum:", kadane(arr))
print(kadane(arr1))
```
