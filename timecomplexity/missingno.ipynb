```python
def find_missing(arr):
    n = len(arr)

    expected_sum = n * (n + 1) // 2
    actual_sum = sum(arr)

    return expected_sum - actual_sum


arr = [0,1,3,4,5,6,7]

result = find_missing(arr)

print("Missing Number =", result)
```
