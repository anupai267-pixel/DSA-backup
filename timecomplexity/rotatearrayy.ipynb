Rotate Array (Right Rotation)

```python
def rotate_right(arr, k):
    k = k % len(arr)

    return arr[-k:] + arr[:-k]


arr = [1, 2, 3, 4, 5]

result = rotate_right(arr, 2)

print("Rotated Array =", result)
```

Rotate Array (Left Rotation)

```python
def rotate_left(arr, k):
    k = k % len(arr)

    return arr[k:] + arr[:k]


arr = [1, 2, 3, 4, 5]

result = rotate_left(arr, 2)

print("Rotated Array =", result)
```
