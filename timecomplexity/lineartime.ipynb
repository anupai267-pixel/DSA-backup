1. Printing All Elements of a List

```python
numbers=[10,20,30,40]
for num in numbers:
    print(num)
```

2. Finding the Sum of All Elements

```python
numbers=[10,20,30,40]
total = 0
for num in numbers:
    total += num
print(total)
```

3. Finding the Maximum Element

```python
numbers=[10,45,20,90,30]
max=numbers[0]
for num in numbers:
    if num > max:
        max = num
print(max)
```

4. Searching for an Element (Linear Search)

```python
numbers = [10,20,30,40]
target = 40
for num in numbers:
    if num == target:
        print("Found")
        break
    
```

5. Counting Even Numbers

```python
numbers=[1,2,3,4,5]
count=0
for num in numbers:
    if num%2==0:
        count +=1
print(count)
```

6. Creating a New List with Squares

```python
numbers=[1,2,3,4,5,6,7]
squares=[]
for num in numbers:
    squares.append(num * num)
print(squares)
```

7. Traversing a String

```python
text="Python"
for ch in text:
    print(ch)
```
