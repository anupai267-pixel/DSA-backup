1. Print All Pairs of Elements

```python
num=[1,2,3,4]
for i in num:
    for j in num:
         print(i,j)
```

2. Multiplication Table

```python
for i in range(1,4):
    for j in range(1,4):
        print(i*j,end=" ")
    print()
```

3. Bubble Sort (Simple Version)

```python
num=[1,2,3,4,5,8]
for i in range(len(num)):
    for j in range(len(num)-1):
        if num[j]>num[j+1]:
         num[j],num[j+1]=num[j+1],num[j]
print(num)
```

4. Count Pair Combinations

```python
num=[1,2,3,4]
count=0
for i in num:
    for j in num:
        count+=1
print(count)
```

5. Star Pattern

```python
n=4
for i in range(n):
    for j in range(n-1):
        print("*",end=" ")
    print()
```
