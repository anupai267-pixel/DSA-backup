1. Binary Search (Most Common Example)

```python
numbers=[10,20,30,40,50]
target = 50
low=0
high=len(numbers)-1
while low <= high:
    mid=(low+high)//2
    if numbers[mid]==target:
        print("Found")
        break
    elif numbers[mid]<target:
        low=mid+1
    else:
        high=mid-1
```

2. Finding a Number by Repeatedly Dividing by 2

```python
n=32
while n>1:
    print(n)
    n=n//2
```

3. Counting How Many Times a Number Can Be Halved

```python
n=64
count=0
while n>1:
    n=n//2
    count +=1
    print(count)
```

4. Powers of 2

```python
n=64
i=1
while i<n:
    print(i)
    i=i*2
    
```

5. Guessing a Number Game

```python
low = 1
high = 100
target = 73
while low <= high:
    mid = (low+high)//2
    if mid == target:
        print("Found:",mid)
        break
    elif mid<target:
        low=mid+1
    else:
        high=mid-1
```
