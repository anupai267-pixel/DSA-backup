Array

```python
arr = [2, 1, 5, 1, 3, 2]

k = 3

window_sum = sum(arr[:k])
max_sum = window_sum

for i in range(k, len(arr)):
    window_sum = window_sum - arr[i-k] + arr[i]
    max_sum = max(max_sum, window_sum)

print(max_sum)
```

Two Pointer Technique

```python
text = "madam"
left = 0
right = len(text) - 1
while left < right:
    if text[left] != text[right]:
        print("Not Palindrome")
        break
    left +=1
    right -=1
else:
    print("Palindrome")
```

Sliding Window Technique

```python
arr=[2,1,5,1,3,2]
k=3
window_sum = sum(arr[:k])
max_sum=window_sum
for i in range(k,len(arr)):
    window_sum = window_sum - arr[i-k] + arr[i]
    max_sum = max(max_sum,window_sum)
print(max_sum)
```
