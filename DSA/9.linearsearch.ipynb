```python
arr = [10, 20, 30, 40, 50]
key = 30
for i in range(len(arr)):
    if arr[i] == key:
        print("Element found at index", i)
        break
else:
    print("Element not found")
```

```python
arr = [5,15,25,35,45]
key=35
position=-1
for i in range(len(arr)):
    if arr[i]==key:
        position=i
        break
if position !=-1:
        print("Found at index",position)
else:
        print("Not Found")
```
