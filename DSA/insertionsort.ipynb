```python
arr = [5, 2, 8, 1, 3]
n = len(arr)
for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
print(arr)
```

```python
marks = [75, 90, 45, 82, 60]

n = len(marks)

for i in range(1, n):
    key = marks[i]
    j = i - 1

    while j >= 0 and marks[j] > key:
        marks[j + 1] = marks[j]
        j -= 1

    marks[j + 1] = key

print("Sorted Marks:", marks)
```
