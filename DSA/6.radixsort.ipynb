```python
def radix_sort(arr):
    exp = 1
    max_num = max(arr)

    while max_num // exp > 0:
        buckets = [[] for _ in range(10)]

        for num in arr:
            digit = (num // exp) % 10
            buckets[digit].append(num)

        arr = []
        for bucket in buckets:
            arr.extend(bucket)

        exp *= 10

    return arr

arr = [170, 45, 75, 90, 802, 24, 2, 66]

print("Before:", arr)
print("After :", radix_sort(arr))
```
